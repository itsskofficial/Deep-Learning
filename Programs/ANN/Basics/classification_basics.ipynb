```python
#creating a decision boundary
import numpy as np

def decision_boundary(model,X,y):
    """Plots decision boundary created by a model predicting on X"""
    #define boundaries
    x_min,x_max=X[:,0].min() -0.1, X[:,0].max() +0.1
    y_min,y_max=X[:,1].min() -0.1, X[:,1].max() +0.1
    xx,yy=np.meshgrid(np.linspace(x_min,x_max,100),
                      np.linspace(y_min,y_max,1000))
    x_in=np.c_[xx.ravel(),yy.ravel()]
    y_pred=model.predict(x_in)

    if len(y_pred[0])>1:
        print("Doing multiclass classification")
        y_pred=np.argmax(y_pred,axis=1).reshape(xx.shape)
    else:
        print("Doing binary classification")
        y_pred=np.round(y_pred).reshape(xx.shape)

    plt.contourf(xx,yy,y_pred,cmap=colormaps.plasma, alpha=0.7)
    plt.scatter(X[:,0], X[:,1], c='y', s=10, cmap=colormaps.RdYlBu)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
```

```python
#check predictions boundary
decision_boundary(model_3,X,y)
```

```python
#finding the best learning rate
tf.random.set_seed(100)

model_6=tf.keras.Sequential()
model_6.add(tf.keras.layers.Dense(100))
model_6.add(tf.keras.layers.Dense(10,activation='relu'))
model_6.add(tf.keras.layers.Dense(1,activation='sigmoid'))

model_6.compile(loss=tf.keras.losses.binary_crossentropy,
                optimizer=tf.keras.optimizers.Adam(),
                metrics=['accuracy'])

lr_scheduler=tf.keras.callbacks.LearningRateScheduler(lambda epoch:1e-4*10**(epoch/20))

history_6=model_6.fit(X_train,y_train,epochs=100,callbacks=[lr_scheduler])
```

```python
pd.DataFrame(history_6.history).plot(figsize=(10,7),xlabel='epochs')
```

```python
#plot learning rate vs loss
learning_rates=1e-4 * 10 **(tf.range(100)/20)
plt.figure(figsize=(10,7))
plt.semilogx(learning_rates, history_6.history['loss'])
```
